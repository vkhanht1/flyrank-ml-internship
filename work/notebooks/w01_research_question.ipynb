## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

total_rows = len(df)
declining_pages = df[df["trend_direction"] == "down"]
num_declining = len(declining_pages)
pct_declining = (num_declining / total_rows) * 100

mean_ctr_down = declining_pages["ctr"].mean()
mean_ctr_up = df[df["trend_direction"] == "up"]["ctr"].mean()

print(f"Total pages analyzed: {total_rows:,}")
print(f"Declining pages needing potential refresh: {num_declining:,} ({pct_declining:.2f}%)")
print(f"Mean CTR comparison - Declining pages: {mean_ctr_down:.4f} vs Growing pages: {mean_ctr_up:.4f}")
```
