## 3. Success metric

*One metric you can defend. What number means 'good'?*

### 3. Success Metric: Precision@K (specifically Precision@50)

*   **Metric:** Precision@50 (the proportion of truly declining pages among the top 50 pages recommended by our model).
*   **Why this metric?** SEO agencies and content teams have limited bandwidth; they can only rewrite a small, fixed number of pages per week (e.g., 50 pages). If we recommend 50 pages and 40 of them are actually declining, our Precision@50 is 80%. This ensures highly efficient use of human writing resources, minimizing the waste of False Positives.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

### 5. Why ML Beats a Fixed Rule Here

A static heuristic rule (e.g., *"If CTR drops by more than 2% and age is over 180 days, flag for refresh"*) suffers from several fatal flaws:
1.  **Rigidity & Thresholding:** Fixed thresholds cannot adapt to different client sizes. A 2% drop on a high-volume page is catastrophic, while on a low-volume page, it's just noise.
2.  **Multivariate Interactions:** Heuristics cannot easily weigh 10+ signals together (e.g., balancing impressions, position decay, click momentum, and seasonality).
3.  **Adaptability:** A rule-based system does not learn. An ML model automatically scales, adjusts weight distributions across different clients (holdout-validated), and handles subtle trends that a human engineer cannot write manual nested `if/else` statements for.
